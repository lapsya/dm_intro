# №1
Нарисуйте графики, показывающий количество браков и разводов на душу населения в США за период с 2000 года по 2014. Отобразите легенду, подпишите оси.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

us_marriage_divorce = pd.read_csv('data/us-marriages-divorces-1867-2014.csv')
```

```python
x = np.arange(2000, 2015)

shift = us_marriage_divorce.Year[0]

y1 = us_marriage_divorce.Marriages[x - shift] / us_marriage_divorce.Population[x - shift]
y2 = us_marriage_divorce.Divorces[x - shift] / us_marriage_divorce.Population[x - shift]

plt.figure()

plt.plot(x, y1, color='red', label='Marriages')
plt.plot(x, y2, color='green', label='Divorces')
plt.legend()
plt.xlabel('year')
plt.ylabel('fraction')
plt.show()
```

# №3
Постройте вертикальную столбчатую диаграмму, сравнивающую количество браков и разводов на душу населения в США в 1900, 1950, 2000 годах.

```python
step = 50
years = np.arange(1900, 2001, step)
width = step / 2

mar = us_marriage_divorce.Marriages[years - shift] / us_marriage_divorce.Population[years - shift]
div = us_marriage_divorce.Divorces[years - shift] / us_marriage_divorce.Population[years - shift]

plt.figure()
plt.xticks(years + width, [str(year) for year in years])
plt.bar(years, mar, color='red', width=width)
plt.bar(years + width, div, color='green', width=width)
plt.show()
```

# №4
Постройте горизонтальную столбчатую диаграмму, сравнивающую количество убийств, совершенных актером в кинофильмах.

```python
actor_kills = pd.read_csv('data/actor_kill_counts.csv')
```

```python
actors = np.arange(len(actor_kills))

kills = actor_kills.Count[actors]

plt.figure()
plt.xticks(actors + 0.5, [''.join(map(lambda x: x[0] + '. ', actor_kills.Actor[actor].split())) for actor in actors])
plt.bar(actors, kills)
plt.show()
```

# №6
Нарисуйте круговую диаграмму, показывающую соотношение способов умереть среди римских императоров.

```python
roman_emperors = pd.read_csv('data/roman-emperor-reigns.csv')
```

```python
roman_emperors.head()
```

```python
deaths = roman_emperors['Cause_of_Death'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(deaths.values,
       labels=deaths.index,
       startangle=90,
       autopct='%1.1f%%',
       colors=['r','g','b','y','m','c','orange', 'grey'])
plt.show()
```
